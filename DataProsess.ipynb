```python
import numpy as np
import pandas as pd
```

```python
#蛋白质信息
#df_protein_train = pd.read_csv('df_protein_train.csv',index_col='Protein_ID')
df_protein_train = pd.read_csv('df_protein_train.csv')
#小分子信息
#df_molecule = pd.read_csv('df_molecule.csv',index_col='Molecule_ID')
df_molecule = pd.read_csv('df_molecule.csv')
#蛋白质与小分子亲和力信息
df_affinity_train = pd.read_csv('df_affinity_train.csv')
```

```python
temp = pd.merge(df_affinity_train,df_protein_train,on='Protein_ID')
```

```python
train_data = pd.merge(temp,df_molecule,on='Molecule_ID')
```

```python
train_data_ = train_data.sort_values(by = 'Protein_ID')
```

```python
train_data_.to_csv('train_sorted.csv',index=None)
```
